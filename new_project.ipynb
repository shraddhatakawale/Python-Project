```python
data["Salary"] = pd.to_numeric(data["Salary"], errors="coerce")
```

```python
data["City"] = data["City"].fillna(data["City"].mode()[0])
data["City"] = data["City"].astype("string")
```

```python
data["Last_Promotion_Date"] = pd.to_datetime(
    data["Last_Promotion_Date"], errors="coerce"
)
```

```python
data["Last_Promotion_Date"] = (
    data.groupby("Department")["Last_Promotion_Date"]
        .transform(lambda x: x.fillna(x.median()))
)
```

```python
data.isnull().sum()
```

```python
data["State"] = data["State"].fillna(data["State"])
```

```python
data["State"].astype("string")
```

```python
data["Active_Project_Name"] =data["Active_Project_Name"].fillna(data["Active_Project_Name"].mode()[0])
```

```python
data["Phone_Number"] = data["Phone_Number"].fillna("unknown")
```

```python
data["Email"] = data["Email"].fillna("Not_Provided")
```

```python
data.isnull().sum()
```

```python
data = data.drop('Performance_Rating' , axis=1)
```

```python
data.rename(columns={'Department' : 'Departments'},inplace=True)
```

```python
data["Gender"] = data["Gender"].replace('Other'  , "Female").astype("string")
```

##  **In Which Department Is The Employee Count Highest ?**

---> According to the Department Wise Employee Count chart show the Sales Department has the highest number of Employee count ,That indicating the strong focus on revenue generation and customer acqusition. also relatively Marketing and Finance has high employee count. Department like HR and R&D have the lowest Workforce representation , suggesting that these department operate with leaner team. This means company mainly focuses on sales and marketing.

```python
plt.figure(figsize=(15,6))

plt.title("Department Wise Employee Count ", family="Georgia",fontsize=15,fontweight="bold")

d1 = data.groupby("Departments", as_index=False)["Employee_ID"].count().sort_values(by="Employee_ID",ascending=False)

E1=['#2ecc71', '#58d68d', '#82e0aa', '#abebc6', '#d5f5e3']

c1 = sns.barplot(x="Departments", y="Employee_ID" , data=d1, palette=E1, edgecolor='#2ecc71',linewidth=1.5)

for i in c1.containers:
    c1.bar_label(i)

plt.xlabel(" ")
plt.ylabel("")

plt.show()
```

##  **Experience Year Wise JoinThe  Date Analysis**

--->The chart show the joing date of employee based on their experience levels. Most employees 0-3 years of experiencejoin recently , while only 5 employee have a 10+ Years of experience. This Indicate the latgey young workforce in the organization.

## **Which gender has a higher count overall?**

---> Overall, females (F) have a higher count than males (M). The Sales Department has the highest female count of 131, while the Marketing Department shows the next highest count with 160 females. The chart shows that females dominate across most Departments, especially in the Sales and Marketing regions. In contrast, males have comparatively lower counts in all Departments, indicating a stronger female representation in the dataset


```python
plt.figure(figsize=(15,6))

plt.title("Departments Wise Gender Count" , family="Georgia" , fontsize="15", fontweight="bold")

E3=['#8e44ad', '#a569bd', '#bb8fce', '#d2b4de', '#ebdef0']
 
plt.grid()

c3 = sns.countplot(x="Gender" , data=data , hue="Departments", palette=E3 , edgecolor='#8e44ad',linewidth=1.5)

for i in c3.containers:
    c3.bar_label(i)

plt.xlabel(" ")
plt.ylabel(" ")
plt.show()
```
